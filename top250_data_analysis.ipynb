```python
import pandas as pd

top250_df = pd.read_csv("top250.csv")
top250_df
```

## 电影上映年代

```python
import matplotlib.pyplot as plt
import operator
%matplotlib inline

plt.hist(top250_df["year"])
plt.show()

year_dict = {}
for year in top250_df["year"]:
    if year in year_dict:
        year_dict[year] += 1
    else:
        year_dict[year] = 1

year_most = max(year_dict, key = year_dict.get)
print("上榜电影数目最多的是{}年，共计{}部".format(year_most, year_dict[year_most]))

sorted_year_dict = sorted(year_dict.items(), key=operator.itemgetter(1), reverse=True)
eight_most_year = sorted_year_dict[:8]
eight_most_year
```

## 电影评分

```python
plt.hist(top250_df["rating_score"])
plt.show()
```

```python
import numpy as np

rating_mean = top250_df["rating_score"].mean()
rating_median = top250_df["rating_score"].median()
rating_stdev = np.std(top250_df["rating_score"])

print("电影评分的平均值为%.2f，中位数为%.2f，标准差为%.4f"%(rating_mean, rating_median, rating_stdev))
```

```python
most_year_rating_mean = {}

for item in eight_most_year:
    year_df = top250_df[top250_df["year"] == item[0]]
    most_year_rating_mean[item[0]] = year_df["rating_score"].mean()
    
most_year_rating_mean
```

```python
movie_of_1994 = top250_df[top250_df["year"] == 1994]
movie_of_1994
```

通过计算比较数量最多的8个年份的电影评分的平均值，可以看出，1994年的电影不仅在榜单中数量很多，并且平均评分也最高。总共12部中，就有5部电影评分不低于9.0分，其中榜单中的最高评分电影就是1994年的《肖申克的救赎》。

```python
rating_over_9_df = top250_df[top250_df["rating_score"] >= 9.0]
plt.hist(rating_over_9_df["rating_score"])
plt.show()
```

## 评价人数

```python
plt.hist(top250_df["rating_people"])
plt.show()
```

```python
rating_people_mean = top250_df["rating_people"].mean()
rating_people_median = top250_df["rating_people"].median()
rating_people_stdev = np.std(top250_df["rating_people"])

print("电影评分人数的平均值为%d，中位数为%d，标准差为%.2f"%(rating_people_mean, rating_people_median, rating_people_stdev))
```

```python
rating_people_over_20w = top250_df[top250_df["rating_people"] > 100000].shape[0]
rate = rating_people_over_20w / top250_df.shape[0]
print("250部电影中单部电影评价人数超过10万的有%d部，占比达到%.1f%%"%(rating_people_over_20w, rate*100))
```

## 电影时长

```python
plt.hist(top250_df["runtime"])
plt.show()
```

```python
# find the longest movie
longest_movie = top250_df.loc[top250_df["runtime"].idxmax()]

# find the shortest movie
shortest_movie = top250_df.loc[top250_df["runtime"].idxmin()]

print(longest_movie, shortest_movie)
```

在榜单中，时长最长的电影是《乱世佳人》，为238分钟，最短的为《萤火之森》，只有45分钟。

```python
runtime_mean = top250_df["runtime"].mean()
runtime_median = top250_df["runtime"].median()
runtime_stdev = np.std(top250_df["runtime"])

print("电影时长的平均值为%d分钟，中位数为%d分钟，标准差为%.2f"%(runtime_mean, runtime_median, runtime_stdev))
```

```python
plt.scatter(top250_df["runtime"], top250_df["rating_score"])
plt.show()

plt.scatter(top250_df["runtime"], top250_df["rating_people"])
plt.show()
```

```python
from scipy.stats import pearsonr

r_rt_rs, p_value = pearsonr(top250_df["runtime"], top250_df["rating_score"])
r_rt_rp, p_value = pearsonr(top250_df["runtime"], top250_df["rating_people"]) 

print(r_rt_rs, r_rt_rp)
```

从散点图和皮尔逊相关系数值来看，榜单中电影的评分与评价人数和电影的时长没有很强的相关性。
