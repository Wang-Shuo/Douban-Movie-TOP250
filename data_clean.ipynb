```python
import pandas as pd

df_1 = pd.read_csv("top250_basic.csv")
df_2 = pd.read_csv("top250_detail.csv")
```

```python
df_1
```

```python
df_1.drop(df_1.columns[0], axis=1, inplace=True)
df_2.drop(df_2.columns[0], axis=1, inplace=True)
```

```python
df_1
```

```python
df_2
```

```python
df = pd.merge(df_1, df_2, right_index=True, left_index=True)
```

```python
df
```

```python
df.drop(["title_y", "star_score", "rating_people"], axis=1, inplace=True)
```

```python
df.rename(columns = {"title_x": "title", "star_num": "rating_people"}, inplace=True)
```

```python
df.set_index("title", inplace=True)
```

```python
df["year"] = df["year"].fillna(0).astype(int)
```

```python
cols = df.columns.tolist()
```

```python
cols
```

```python
new_cols = ["year"] + ["rating_score"] + ["rating_people"] + ["runtime"] + ["star5"] + ["star4"] + ["director"] + ["screenwriter"] + cols[2:5] + ["country"] + cols[7:9] + ["imdb_link"] + ["quote"]
```

```python
df = df[new_cols]
```

```python
df
```

```python
df.dropna(thresh = 10, inplace= True)
```

```python
import re

def clean_runtime_1(runtime):
    return re.findall(r"\d+", str(runtime))

    
df["runtime"] = df["runtime"].apply(clean_runtime_1)

def clean_runtime_2(runtime):
    if runtime != []:
        return runtime[0]
    else:
        return 81
    
df["runtime"] = df["runtime"].apply(clean_runtime_2)
        
```

```python
df
```

```python
df.to_csv("cleaned_output.csv")
```
